```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
```

```python
#6.1
small_cap = 'D:\intern\module 1\Small_Cap'
large_cap = 'D:\intern\module 1\Large_Cap'
mid_cap = 'D:\intern\module 1\Mid_Cap'
```

```python
list1 = os.listdir(small_cap)
list2 = os.listdir(mid_cap)
list3 = os.listdir(large_cap)
```

```python
print(list1[:3], list2[:3], list3[:3])
```

```python
random.shuffle(list1)
random.shuffle(list2)
random.shuffle(list3)
```

```python
list1 = list1[:10]
list2 = list2[:10]
list3 = list3[:10]
```

```python
li = []
for i in list1:
    df = pd.read_csv(small_cap+'\\'+i)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    df = df[df['Series']=='EQ']
    df[i[:-4]] = df['Close Price']
    li.append(df[i[:-4]])
```

```python
for i in list2:
    df = pd.read_csv(mid_cap+'\\'+i)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    df = df[df['Series']=='EQ']
    df[i[:-4]] = df['Close Price']
    li.append(df[i[:-4]])
```

```python
for i in list3:
    df = pd.read_csv(large_cap+'\\'+i)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    df = df[df['Series']=='EQ']
    df[i[:-4]] = df['Close Price']
    li.append(df[i[:-4]])
```

```python
df_30stocks = pd.concat(li, axis = 1, ignore_index = False)
```

```python
#6.2
df_30stocks = df_30stocks[-252:]
```

```python
annual_returns = df_30stocks.pct_change()
```

```python
annual_returns.shape
```

```python
annual_returns.head()
```

```python
annual_returns.tail()
```

```python
annual_returns_mean = annual_returns.mean()
annual_returns_vol = annual_returns.std()
```

```python
print("Annual Returns Mean")
print(annual_returns_mean)
```

```python
print("Annual Returns std")
print(annual_returns_vol)
```

```python
labels = list(annual_returns_mean.index)
X = np.array([annual_returns_mean, annual_returns_vol])
X = X.T
```

```python
print(X)
```

```python
#6.3
from sklearn.cluster import KMeans
```

```python
cost = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(X)
    cost.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), cost)
plt.title("Elbow Method")
plt.xlabel("No. of Clusters")
plt.ylabel("Cost")
plt.show()
```

From above graph we can identify four are optimal no. of clusters

```python
kmeans = KMeans(n_clusters = 4, random_state = 0)
y_labels = kmeans.fit_predict(X)
```

```python
print(y_labels)
print(y_labels.shape)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X[y_labels==0, 0], X[y_labels==0, 1], color = 'red', label = "Cluster 1")
plt.scatter(X[y_labels==1, 0], X[y_labels==1, 1], color = 'yellow', label = "Cluster 2")
plt.scatter(X[y_labels==2, 0], X[y_labels==2, 1], color = 'magenta', label = "Cluster 3")
plt.scatter(X[y_labels==3, 0], X[y_labels==3, 1], color = 'green', label = "Cluster 4")
plt.xlabel('Annual Returns Mean')
plt.ylabel('Annual Returns Volatality')
plt.title('Clusters of Stocks')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X[y_labels==0, 0], X[y_labels==0, 1], color = 'red', label = "Cluster 1")
plt.scatter(X[y_labels==1, 0], X[y_labels==1, 1], color = 'yellow', label = "Cluster 2")
plt.scatter(X[y_labels==2, 0], X[y_labels==2, 1], color = 'magenta', label = "Cluster 3")
plt.scatter(X[y_labels==3, 0], X[y_labels==3, 1], color = 'green', label = "Cluster 4")
for j in range(0,30):
    plt.annotate(labels[j], (X[j,0], X[j,1]))
plt.xlabel('Annual Returns Mean')
plt.ylabel('Annual Returns Volatality')
plt.title('Clusters of Stocks')
plt.legend()
plt.show()
```

```python
#6.4
c1 = []
c2 = []
c3 = []
c4 = []
```

```python
e = 0
```

```python
for k in y_labels:
    if k==0 :
        c1.append(labels[e])
        e = e+1
    elif k==1 :
        c2.append(labels[e])
        e = e+1
    elif k==2:
        c3.append(labels[e])
        e = e+1
    else:
        c4.append(labels[e])
        e = e+1
```

```python
print(c1)
print(c2)
print(c3)
print(c4)
```

```python
clusters = pd.DataFrame({'Cluster 1':pd.Series(c1),
                         'Cluster 2':pd.Series(c2),
                         'Cluster 3':pd.Series(c3),
                         'Cluster 4':pd.Series(c4)})
```
