```python
#1.4
stock_data['year'] = stock_data['Date'].dt.year
stock_data['month'] = stock_data['Date'].dt.month
```

```python
#converting M to 10^6
temp = stock_data['Total Traded Quantity'].values.tolist()
for i in range(len(temp)):
    temp[i] = temp[i][:][:-1]
    temp[i] = float(temp[i])*pow(10,6)
```

```python
temp
```

```python
stock_data['Total Traded Quantity'] = temp
```

```python
stock_data.head()
```

```python
stock_data["price*volume"] = stock_data["Close Price"]*stock_data["Total Traded Quantity"]
```

```python
grouped_by_month_year = stock_data.groupby(['year','month']).sum()
```

```python
vwap = np.array(grouped_by_month_year['price*volume']/grouped_by_month_year['Close Price'])
```

```python
#1.5
def avg_last_n_days(N, series):
    return (series[:N].sum())/N

def pf_ls_n_days(N, series):
    return ((series[0] - series[N-1])/series[N-1])*100
```

```python
#Average Values of last N days
print("1 week average: "+str(avg_last_n_days(7, stock_data['Close Price'])))
print("2 weeks average: "+str(avg_last_n_days(14, stock_data['Close Price'])))
print("1 month average: "+str(avg_last_n_days(30, stock_data['Close Price'])))
print("3 months average: "+str(avg_last_n_days(90, stock_data['Close Price'])))
print("6 months average: "+str(avg_last_n_days(180, stock_data['Close Price'])))
print("1 year average: "+str(avg_last_n_days(365, stock_data['Close Price'])))
```

```python
#profit\loss of last N days -ve profit implies loss
print("1 week profit: "+str(pf_ls_n_days(7, stock_data['Close Price'])))
print("2 weeks profit: "+str(pf_ls_n_days(14, stock_data['Close Price'])))
print("1 month profit: "+str(pf_ls_n_days(30, stock_data['Close Price'])))
print("3 months profit: "+str(pf_ls_n_days(90, stock_data['Close Price'])))
print("6 months profit: "+str(pf_ls_n_days(180, stock_data['Close Price'])))
print("1 year profit: "+str(pf_ls_n_days(365, stock_data['Close Price'])))
```

```python
#1.6
stock_data['Day_Perc_Change'] = stock_data['Close Price'].pct_change() * 100
stock_data.at[0,'Day_Perc_Change'] = 0
stock_data.head()
```

```python
#1.7
for i in range(len(stock_data['Close Price'])):
    param = stock_data['Day_Perc_Change'][i]
    if -0.5<= param <= 0.5:
        stock_data.at[i,'Trend'] = 'Slight or No Change'
    elif 0.5<param<=1:
        stock_data.at[i,'Trend'] = 'Slight positive'
    elif -1<=param<-0.5:
        stock_data.at[i,'Trend'] = 'Slight negative'
    elif 1<param<=3:
        stock_data.at[i,'Trend'] = 'Positive'
    elif -3<=param<-1:
        stock_data.at[i,'Trend'] = 'Negative'
    elif 3<param<=7:
        stock_data.at[i,'Trend'] = 'Among top gainers'
    elif -7<=param<-3:
        stock_data.at[i,'Trend'] = 'Among top losers'
    elif param>7:
        stock_data.at[i,'Trend'] = 'Bull run'
    elif param<-7:
        stock_data.at[i,'Trend'] = 'Bear drop'
```

```python
#1.8
#Mode
print("Mode Values")
stock_data.groupby(['Trend'])['Total Traded Quantity'].mean()
```

```python
#Median
print("Median Values")
stock_data.groupby(['Trend'])['Total Traded Quantity'].median()
```
